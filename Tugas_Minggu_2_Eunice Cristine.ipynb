Generate Data

```python
import time
import pandas as pd
import numpy as np

start = time.time()

# Misal skema dataset-nya seperti berikut: => Silahkan dimodifikasi sesuai case Anda
kolomFitur_X_plus_Target_Y = ['JenisAsset','MerkAsset', 'TahunPembelian', 'SistemOperasi', 'Storage', 'Prosesor', 'TargetMaintenance']

# Interval atau Variasi nilai fitur
fitur_jenis_aset = ['Komputer','Laptop']
fitur_merk_asset = ['Asus','Dell','Acer','Toshiba']
fitur_tahun_pembelian = [2014,2021]
fitur_sistem_operasi = ['Windows7','Windows8','Windows10','Windows11']
fitur_jenis_storage = ['HDD','SSD']
fitur_prosesor = ['Intel', 'AMD']
labelTargetY = [0, 7, 2] #dalam bulan

# set bykData = 3*np.power(10,7)
bykData = 10000000
bykFitur = len(kolomFitur_X_plus_Target_Y)-1

print('Banyak Data = ', bykData)
print('Banyak Fitur = ', bykFitur)

# generate isi dataset
content_dataGenerate = np.array([np.arange(bykData)]*(bykFitur+1)).T
df_gen = pd.DataFrame(content_dataGenerate, columns=kolomFitur_X_plus_Target_Y)

# set secara random nilai Fitur untuk generate data
df_gen ['JenisAsset'] = np.random.choice(fitur_jenis_aset, df_gen.shape[0])
df_gen ['MerkAsset'] = np.random.choice(fitur_merk_asset, df_gen.shape[0])
df_gen ['TahunPembelian'] = np.random.randint(fitur_tahun_pembelian[0], fitur_tahun_pembelian[1], df_gen.shape[0])
df_gen ['SistemOperasi'] = np.random.choice(fitur_sistem_operasi, df_gen.shape[0])
df_gen ['Storage'] = np.random.choice(fitur_jenis_storage, df_gen.shape[0])
df_gen ['Prosesor'] = np.random.choice(fitur_prosesor, df_gen.shape[0])
df_gen ['TargetMaintenance'] = np.random.randint(labelTargetY[0], labelTargetY[1], df_gen.shape[0])

# save dataframe generate ke *.csv
file_name_data_generate = 'Skema_Dataset_Eunice.csv'
df_gen.to_csv(file_name_data_generate, encoding='utf-8', index=False)

# read file *.csv dan tampilkan
data_generate = pd.read_csv(file_name_data_generate)

# Menampilkan data
print('\nMenampilkan hasil generate dataset:')
display(data_generate)

end = time.time()
print(end - start)
```

Regresi 30 Data

```python
df = pd.read_csv('Skema_Dataset_Eunice_30.csv', usecols=['JenisAsset','MerkAsset', 'TahunPembelian', 'SistemOperasi', 'Storage', 'Prosesor', 'TargetMaintenance'])
df
```

```python
df['JenisAsset']=df['JenisAsset'].astype('category').cat.codes
df['MerkAsset']=df['MerkAsset'].astype('category').cat.codes
df['SistemOperasi']=df['SistemOperasi'].astype('category').cat.codes
df['Storage']=df['Storage'].astype('category').cat.codes
df['Prosesor']=df['Prosesor'].astype('category').cat.codes
```

```python
df
```

```python
X = df.drop('TargetMaintenance', axis=1).values
y = df['TargetMaintenance'].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
```

```python
start_time = time.time()

lin_reg.fit(X_train, y_train)

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
start = time.time()

akurasi = lin_reg.score(X_test, y_test)
print("akurasi : ", end="")
print(akurasi)

end = time.time()
waktu = end - start
print("waktu : ", end="")
print(waktu)
```
